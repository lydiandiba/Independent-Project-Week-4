```
# importing pandas and numpys
import pandas as pd

import numpy as np
```

```
# loading datasets
cars = pd.read_csv("Autolib_dataset .csv")

# previewing datasets
cars
```

```
# accessing information from our dataset
cars.info()
```

```
# dropping the column names that we wont be needing in analysis
cars.drop(['Cars','Displayed comment','Station type','Charge Slots','Charging Status','Scheduled at'], axis = 1, inplace= True)
cars
```

```
#looking for missing values
cars.isnull().any()
```

```
#merging the columns year, month and day to form one column named date

cars['Date'] = cars["year"].map(str) +':'+ cars["month"].map(str) +':'+ cars["day"].map(str)
cars.head()
```

```
#merging the columns hour and minute to get a column time
cars['Time'] = cars["hour"].map(str) +':'+ cars["minute"].map(str) 
cars.head()
```

```
#merging our newly formed columns date and time to form a column date_time
cars['Date_Time'] = cars["Date"].map(str) +':'+ cars["Time"].map(str)
cars.head()
```

```
#merging date, time and date_time
cars['Datetimecomb']= cars["Date"].map(str) +':'+ cars["Time"].map(str) +':'+cars["Date_Time"].map(str)
cars.head()
```

```
from datetime import datetime
cars_1 = cars
cars_1['Datetimecomb'] = pd.to_datetime(cars['Datetimecomb'] , format='%Y:%m:%d:%H:%M:%S', errors = 'coerce')
cars_1.info()
```

```
#fixing the messy columns
cars.columns = cars.columns.str.strip().str.lower().str.replace(' ', '_').str.replace('.', '')
cars.head()
```

```
#creating new columns for the usage of the bluecars, utilib and utilib 1.4
cars['usage_bluecar'] = cars.bluecar_counter.diff()
cars.head()
```

```
cars['usage_utilib'] = cars.utilib_counter.diff()
cars.head()
```

```
#creating the usage_utilib14
cars['usage_utilib_14'] = cars.utilib_14_counter.diff()
cars.head()
```

```
#from our new usage columns;
#all the positive values represent cars that were returned
#the negative values represent the cars that were picked 
#and zero values mean nothing happened, no cars were picked or dropped 
```

```
#viewing our bluecar_counter and the usage_counter
cars[['bluecar_counter','usage_bluecar']]
```

```
#Identify the most popular hour of the day for picking up a shared electric car (Bluecar) in the city of Paris over the month of April 2018
cars[(cars.city == 'Paris')&
     (cars.usage_bluecar < 0)].groupby('hour')['hour'].count().sort_values(ascending = False).head(1)
```

Most popular hour for picking up a bluecar is 2100hours

```
#we can also find the most popular hour by subseting
paris_df = cars[cars.city =='Paris']
paris_df[['bluecar_counter','usage_bluecar','hour']]
paris_df
```

```
paris_df[(paris_df.usage_bluecar < 0)].groupby('hour')['hour'].count().sort_values(ascending = False). head(1)
```

```
#What is the most popular hour for returning cars?
cars[(cars.city == 'Paris') & 
   (cars.usage_bluecar > 0)].groupby('hour')['usage_bluecar'].count().sort_values(ascending = False).head(1)
```

Most popular hour for returning cars is 0300 hours

```
#What station is the most popular overall?
cars[(cars.kind == 'STATION')& (cars.status == 'ok')].groupby('public_name')[['kind']].count().sort_values(by = 'kind', ascending = False).head(1)
```
